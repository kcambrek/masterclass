### Batch-effect exercise

In statistics, quantile normalization is a technique for making two distributions identical in statistical properties.
Clustering techniques will see the batch effect as an seperation of the data

```python
# Libraries
# %reset -f
import sys, os
sys.path.append(os.path.join(os.getcwd(), "py_scripts"))
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
# from sklearn.datasets.samples_generator import make_blobs, make_circles
# from scipy.cluster.hierarchy import inconsistent

import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# from clusteval import clusteval
# from silhouette import silhouette
# from dbindex import dbindex
# from scatter import scatter
# from DBSCAN import DBSCAN
# from HDBSCAN import HDBSCAN
from dendrogram_fancy import dendrogram_fancy
from distfit import distfit
from hist import hist
from imagesc import imagesch

from sklearn.datasets.samples_generator import make_blobs, make_circles
from sklearn.preprocessing import quantile_transform
from tsneBH import tsneBH


# some setting for this notebook to actually show the graphs inline
#%matplotlib inline
#np.set_printoptions(precision=5, suppress=True)  # suppress scientific float notation
```

```python
# Make random data sets with 3 classes
[X, labels_true] = make_blobs(n_samples=1000, centers=3, n_features=100)
```

```python
# Make random data sets with 3 classes
[X, labels_true] = make_blobs(n_samples=1000, centers=3, n_features=100)
# Dimensionality reduction to view the sample-sample relationships
Xtsne = tsneBH(X, showprogress=0)
# Show the raw data
_=scatter(Xtsne[:,0],Xtsne[:,1], size=100, labx=labels_true, title='Raw data', width=5, height=5)
```

```python
# Idea: You received data sets from two different parties/companies.
# Include batch-effect at every class
# 
X[np.where(labels_true==1)[0][0:100],:] = X[np.where(labels_true==1)[0][0:100],:]+5
X[np.where(labels_true==2)[0][0:100],:] = X[np.where(labels_true==2)[0][0:100],:]+5
X[np.where(labels_true==0)[0][0:100],:] = X[np.where(labels_true==0)[0][0:100],:]+5

# Scatter the batch effect
XtsneBatch = tsneBH(X, showprogress=0)
scatter(XtsneBatch[:,0],XtsneBatch[:,1], size=100, labx=labels_true, title='Raw data with batch-effect', width=5, height=5)
```

```python
# Quantile normalization to remove batch effect
Xquantile     = quantile_transform(X, axis=1)
XtsneQuantile = tsneBH(Xquantile, showprogress=0)
# Scatter the quantile normalized data
scatter(XtsneQuantile[:,0],XtsneQuantile[:,1], size=100, labx=labels_true, title='Quantile normalized data', width=5, height=5)
```

### EXERCISE 15
The distributions is exponential.<br>
Note that the features (genes) are not independent, but that expression levels can be highly correlated.
<br>
#### How would you proceed before clustering the data?
<br><b>Step 1: Ignore the data distribution and simply cluster as is</b>
<b><br>Step 2: Transform the data in such a way that it follows gaussian (normal-like distribution)</b>
<b><br>Step 3: Transform the data (as in step 2), then remove features by feature-selection</b>

<i> For both cases, use the most suitable distance metric and linkage type</i>
<br>https://docs.scipy.org/doc/scipy/reference/generated/scipy.spatial.distance.pdist.html

```python
# Step 3
v
# RAW data
data = df.copy()
hist(data.values.ravel(), width=5, height=5, title='RAW')

# 1. Transform the data by log2 but first add pseudocount to prevent log2(0)
dataL = np.log2(data+1)
hist(dataL.values.ravel(), width=5, height=5, title='Log2 transformation')

# The data looks skewed because of the log2(1)=0 values
# 2. Apply zero-mean normalization, per feature over all samples
dataLZ = dataL.copy()
rowmeans=np.mean(dataL, axis=1)
for i in range(dataLZ.shape[0]):
    dataLZ.iloc[i,:] = dataL.values[i,:] - rowmeans[i]
#end
hist(dataLZ.values.ravel(), width=5, height=5, title='Log2->Zero Mean')
#distfit(dataLZ[:,0])

# 3. Feature selection on variance
minvar=8
featvar=dataLZ.var(axis=1)
hist(featvar, width=5, height=5, title='Variance across features')
dataLZF = dataLZ.iloc[np.where(featvar>=minvar)[0],:]
```

```python
dataLZF.shape
```

```python
# Cluster the features
Z=linkage(dataLZF.values, metric='correlation', method='complete')
plt.figure(figsize=(25, 10))
plt.clf()

max_d=0.8
out=dendrogram_fancy(Z, labels=dataLZF.index.values, leaf_rotation=90, leaf_font_size=15, show_contracted=True, annotate_above=1.25, max_d=max_d)
labxNew = fcluster(Z, max_d, criterion='distance')

#plt.figure(figsize=(25, 10))
#for i in range(0,len(idx)):
#    plt.plot(dataLZF.values[idx[i],:], alpha=0.6)
#end
#plt.show()
```

```python
# Make dendrogram
# plt.figure(figsize=(25, 10))
# Z=linkage(dataLZF.T, method='complete', metric='correlation')
# d=dendrogram(Z,leaf_font_size=8, labels=metadata.labx.values)

#Z = linkage(dist_mat, "complete")
Z=linkage(dataLZF.values.T, metric='correlation', method='complete')

plt.figure(figsize=(25, 10))
plt.clf()

#dendrogram(Z, leaf_font_size=15, labels=metadata.labx.values)
#d=dendrogram(Z,leaf_font_size=8, labels=metadata.labx.values)
out=dendrogram_fancy(Z, labels=metadata.labx.values, leaf_rotation=90, leaf_font_size=15, show_contracted=True, annotate_above=1.1, max_d=1.1)

# Get colors per labx
getcolors  = sns.color_palette('Pastel1',len(metadata.labx.unique())).as_hex()
my_palette = dict(zip(df.columns.unique(),getcolors))
row_colors = metadata.labx.map(my_palette)

# Assignment of colors to Text-labels
ax = plt.gca()
xlbls = ax.get_xmajorticklabels()
for lbl in xlbls:
    lbl.set_color(my_palette[lbl.get_text()])
#end

plt.title('Dendrogram')
plt.xlabel('Patients')
plt.ylabel('Correlation distances')
plt.show()
```

### EXERCISE 18
How would you cluster the data when it is catagorized?

```python
distout=distfit(dataLZF.values, distribution='norm',alpha=0.95, bins=100)
# Make discrete values
dataLZFD = np.zeros(dataLZF.shape, dtype='int')
I1=dataLZF.values>=distout[0]['CII_min_alpha']
I11=np.logical_and(dataLZF.values<=distout[0]['CII_max_alpha']-2,dataLZF.values<=distout[0]['CII_max_alpha'])
I2=dataLZF.values<=distout[0]['CII_max_alpha']
I22=np.logical_and(dataLZF.values<=distout[0]['CII_max_alpha']+2,dataLZF.values<=distout[0]['CII_max_alpha'])

# Set values at one or 
dataLZFD[I1]=2 # Upregulation
dataLZFD[I11]=1 # Upregulation
dataLZFD[I22]=-1 # Upregulation
dataLZFD[I2]=-2 # Downregulation
dataLZFD = pd.DataFrame(dataLZFD, index=dataLZF.index, columns=dataLZF.columns)
dataLZFD.head(10)
```
